We need to find the slope and the y-intercept of a straight line that minimizes the errors... Let's generalize

Define cloud point data $$(x_i,y_i)$$

Define the best fit line  \begin{equation} \hat{y_i} = a + Bx_i \tag{1} \end{equation}


Where we need to minimize the sum of squared errors (SSE) function \begin{equation} S =  \sum \limits _{i=1} ^{n} (y_i - \hat{y_i})^2  \tag{2}  \end{equation}


Subsutitute $\hat{Y_i}$ in (2) and we get the OLS function : \begin{equation} f(a,B) =  \sum \limits _{i=1} ^{n} (y_i - a - Bx_i)^2 \tag{3}  \end{equation}

Quadratic ...minimize the function in (3) by taking the first derrivate equal to zero 

Partial first order deivative	: $$\frac{\partial f}{\partial a} [ \sum \limits _{i=1} ^{n} (y_i - a - Bx_i)^2 ] = 0 $$

Chain Rule :  $$ \Rightarrow \sum \limits _{i=1} ^{n} -2(y_i - a - Bx_i)  = 0 $$

$$  \Rightarrow \sum \limits _{i=1} ^{n} y_i    -  \sum \limits _{i=1} ^{n} a   -B  \sum  \limits _{i=1} ^{n} x_i = 0 $$

$$  \Rightarrow \sum \limits _{i=1} ^{n} y_i    -  na   -B  \sum  \limits _{i=1} ^{n} x_i = 0 $$

$$  \Rightarrow a = \frac{\sum _{i=1} ^{n} y_i -B  \sum_{i=1} ^{n} x_i}{n} $$

 By the definition of the mean \begin{equation} \therefore \boxed{a = \bar{y} -B\bar{x} } \tag{4}  \end{equation} 

# Find the slope  B:

Substitute a from (4) in the OLS function : \begin{equation} \Rightarrow \sum \limits _{i=1} ^{n} (y_i - a - Bx_i)^2  \end{equation}
\begin{equation} \Rightarrow \sum \limits _{i=1} ^{n} (y_i - (\bar{y} -B\bar{x}) - Bx_i)^2   \end{equation}
Rearrange \begin{equation} \Rightarrow \sum \limits _{i=1} ^{n} [ (y_i -\bar{y}) -B(x_i -\bar{x}) ]^2  \end{equation}

Partial first order derrivative wrt B  \begin{equation} \frac{\partial f}{\partial B}  \sum \limits _{i=1} ^{n} [ (y_i -\bar{y}) -B(x_i -\bar{x}) ]^2 = 0   \end{equation}

Chain rule \begin{equation} \Rightarrow -2 \sum \limits _{i=1} ^{n} [ (y_i -\bar{y}) -B(x_i -\bar{x})] (x_i -\bar{x}) = 0  \end{equation}

Expand \begin{equation} \Rightarrow -2 \sum \limits _{i=1} ^{n} (y_i -\bar{y})(x_i -\bar{x}) -B(x_i -\bar{x})^2 = 0   \end{equation}

eliminate 2 and split terms \begin{equation} \Rightarrow  \sum \limits _{i=1} ^{n} (y_i -\bar{y})(x_i -\bar{x}) -B\sum \limits _{i=1} ^{n} (x_i -\bar{x})^2 = 0  \end{equation}

solve for B  \begin{equation} \therefore \boxed{B = \frac{\sum _{i=1} ^{n} (y_i -\bar{y})(x_i -\bar{x})}{\sum _{i=1} ^{n} (x_i -\bar{x})^2} }  \tag{5} = \frac{Covariance(x,y)}{Variance(x)} \end{equation}

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({"var1": [1,2,3,4,5],
                   "var2": [7,14,15,18,19]
                   })

xmean = df["var1"].mean()
ymean = df["var2"].mean()
df['xxmean'] = df['var1'] - xmean
df['yymean'] = df['var2'] - ymean
df['xxmean*yymean'] = df['xxmean']*df['yymean']
df['xxmean**2'] = df['xxmean']**2
SSxy_numerator = df['xxmean*yymean'].sum()
SSxx_denomenator = df['xxmean**2'].sum()


slope = SSxy_numerator/SSxx_denomenator
intercept = ymean - slope*xmean
print('Intercept using derivation =' , round(intercept, 2))
print('Slope using derivation =', slope)

X = df.iloc[:, 0].values.reshape(-1, 1)  
Y = df.iloc[:, 1].values.reshape(-1, 1)  
lr = LinearRegression()  
lr.fit(X, Y)  
Y_pred = lr.predict(X)  
print('Intercept using sklearn =' ,round(lr.intercept_[0],2))
print('Slope using sklearn =', round(lr.coef_[0][0],2))


print(lr.score(X, Y))



```
